##Crear una función que dados los coeficientes de un polinomio de segundo grado grafiquela parábola en un intervalo dado. Además, marcar las raíces en el gráfico, por ejemplo con puntos rojos y un texto indicando su valor con 3 decimales. Importante, una condición más al problema es que el texto debe escribirse utilizando latex.

Más especificamente, en esta implementación, vamos a considerar la parábola en el intervalo (x ∈ [−4.5, 4]). Además, al marcar las raíces en el gráfico consideramos solo las raíces reales e indico su valor con 3 decimales utilizando latex (ejemplo: xi = 1.22).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bhaskara(a, b=0, c=0):
    det = np.sqrt(b * b - 4 * a * c + 0j) / (2 * a)
    aux = -b / (2 * a)
    return [aux + det, aux - det]


def cuadratica(a, b, c, x):
    return a * x * x + b * x + c


def cuadratica_grafico(a, b, c):
    x = np.linspace(-4.5, 4, 100)
    y = cuadratica(a, b, c, x)
    plt.plot(x, y)
    plt.xlim(-4.5, 4)
    raices_reales = []
    raices = bhaskara(a, b, c)

    for i in raices:
        if np.imag(i) == 0 and np.real(i)!=0:
            raices_reales.append(np.real(i))
        elif np.imag(i)==0 and np.real(i)==0:
            raices_reales=[0]
            break
    if raices_reales != []:
        plt.scatter(raices_reales, np.zeros(len(raices_reales)), color="red", marker='o',s=50)

    for i in range(len(raices_reales)):
        label = r"$X_{0:} = {1:.2f}    $".format(i+1,raices_reales[i])
        plt.annotate(label, xy=(raices_reales[i]-i, 1.5),fontsize=8)
    plt.xlabel(r'Eje X')
    plt.xlim([-4.5,4])

    return

plt.figure(figsize=(10, 5))
ax1=plt.subplot(131)
cuadratica_grafico(1, 1, -1)
ax1.set_ylabel(r'Eje Y')

ax2=plt.subplot(132)
cuadratica_grafico(1,0,0)

ax3=plt.subplot(133)
cuadratica_grafico(1,0,1)

ax1.title.set_text('a=1, b=1, c=-1')
ax2.title.set_text('a=1, b=0, c=0')
ax3.title.set_text('a=1, b=0, c=1')
plt.subplots_adjust(bottom=0.25, top=0.75)
plt.tight_layout()
plt.show()
```
