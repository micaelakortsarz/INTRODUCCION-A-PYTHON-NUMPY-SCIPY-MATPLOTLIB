## 2- Generar datos y cálculo de histogramas, medias y desviaciones estándar

Calcular el histograma, la media (μ) y la desviación estándar (σ) de los datos generados mediante:
data= np.random.gamma(3,2,1000)

Verificar que sean las correctas mediante una visualización del histograma y comprobando que los valores teóricos y empíricos de μ y σ coinciden.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
print("Ejercicio 2")
data = np.random.gamma(3, 2, 1000)
h = np.hstack(data)
mean, std = np.mean(h), np.std(h)
print("La media es {} y la desviación estándar es {}".format(mean, std))

plt.hist(h, 25, facecolor='purple',edgecolor='purple', alpha=0.2)
g=stats.gamma(3,0,2)
x=np.linspace(0,30,100)
plt.axvline(mean)
label =r"Media ={0:.3f}".format(mean)
plt.annotate(label, xy=(mean+std+0.1, 120))
plt.plot(x,1000*g.pdf(x))
label =r"Desviación estándar ={0:.3f}".format(std)
plt.annotate(label, xy=(mean+std+0.1, 110))

plt.show()
```
