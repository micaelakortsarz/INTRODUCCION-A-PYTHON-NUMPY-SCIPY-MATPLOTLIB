```python
import matplotlib.pyplot as plt
import numpy as np
```

Consigna: A partir de la definición de la función y el código de abajo intente obtener un gráfico como
el siguiente teniendo en cuenta el colormap (bone), la interpolación de la imagen y el origen
(pensar que función de numpy necesita para crear X e Y):

```python

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 10
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)
plt.imshow(f(X, Y), cmap="bone", aspect='equal', interpolation=None, vmin=-0.65,
           vmax=1.05, origin='lower', extent=None, filternorm=True, 
           filterrad=4.0, resample=None)

plt.colorbar(shrink=0.8).ax.tick_params(axis='y', direction='in')

plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()

```

A partir de la definición de la función y el código anterior obtener un gráfico  teniendo en cuenta el colormap (hot) y n=256

Ayuda: usar los comandos contourf, contour y clabel.

```python

print("Contorno")
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)
z = f(X, Y)
plt.contourf(z, cmap='hot', alpha=0.75)
c = plt.contour(z, colors='black', alpha=0.75)
plt.clabel(c, inline=1, fontsize='x-small', colors='black')
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()

```

```python

print("Scatter")
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
colors = np.arctan2(Y, X)
plt.scatter(X, Y, s=25, c=colors, cmap='jet', alpha=0.6, edgecolors='black', 
            linewidths=0.75)
plt.xlim(-1.75, 1.75)
plt.ylim(-1.75, 1.75)
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()
```
